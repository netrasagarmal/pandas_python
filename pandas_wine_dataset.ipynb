# Assuming the DataFrame is named `df`
import pandas as pd
import numpy as np

# 1. Mean Alcohol for each Wine category
df.groupby('Wine')['Alcohol'].mean()

# 2. Median Malic.acid where Hue > mean
df[df['Hue'] > df['Hue'].mean()]['Malic.acid'].median()

# 3. Std dev of Phenols per Wine type
df.groupby('Wine')['Phenols'].std()

# 4. High Flavanoids flag + avg Ash
df['High_Flavanoids'] = df['Flavanoids'] > df['Flavanoids'].mean()
df.groupby('High_Flavanoids')['Ash'].mean()

# 5. Acl of wine with max Color.int per Wine group
df.loc[df.groupby('Wine')['Color.int'].idxmax()][['Wine', 'Acl']]

# 6. Bin Alcohol and count per bin and Wine
df['Alcohol_bin'] = pd.cut(df['Alcohol'], bins=[0, 13, 14, np.inf], labels=['Low', 'Medium', 'High'])
df.groupby(['Wine', 'Alcohol_bin']).size()

# 7. Filter Mg < 90, avg Proline per Wine
df[df['Mg'] < 90].groupby('Wine')['Proline'].mean()

# 8. Top 3 Proline rows per Wine
df.sort_values(['Wine', 'Proline'], ascending=[True, False]).groupby('Wine').head(3)[['Alcohol', 'Proline', 'Wine']]

# 9. Mean and range of Phenols per Wine
grouped = df.groupby('Wine')['Phenols']
pd.DataFrame({'mean': grouped.mean(), 'range': grouped.max() - grouped.min()})

# 10. Pivot table: avg and count of Flavanoids per Wine
pd.pivot_table(df, values='Flavanoids', index='Wine', aggfunc=['mean', 'count'])

# 11. TotalPhenols column and group mean
df['TotalPhenols'] = df['Phenols'] + df['Flavanoids'] + df['Nonflavanoid.phenols']
df.groupby('Wine')['TotalPhenols'].mean()

# 12. iterrows: count rows with Hue & OD > median
count = sum((row.Hue > df['Hue'].median()) and (row.OD > df['OD'].median()) for _, row in df.iterrows())

# 13. apply: flag if Ash + Acl + Mg > 250
df['High_Total'] = df.apply(lambda x: x['Ash'] + x['Acl'] + x['Mg'] > 250, axis=1)

# 14. Top row by Malic.acid per Wine (descending)
df.sort_values(['Wine', 'Malic.acid'], ascending=[True, False]).groupby('Wine').head(1)

# 15. Wine type with highest mean OD
df.groupby('Wine')['OD'].mean().idxmax()

# 16. Drop bottom 5% Hue, avg Color.int per Wine
threshold = df['Hue'].quantile(0.05)
df[df['Hue'] > threshold].groupby('Wine')['Color.int'].mean()

# 17. Top 10% Alcohol, group avg Malic.acid
cutoff = df['Alcohol'].quantile(0.9)
df[df['Alcohol'] > cutoff].groupby('Wine')['Malic.acid'].mean()

# 18. Z-score normalize Acl, group mean
df['Acl_z'] = (df['Acl'] - df['Acl'].mean()) / df['Acl'].std()
df.groupby('Wine')['Acl_z'].mean()

# 19. Id (index) with min Proanth per Wine
df.groupby('Wine')['Proanth'].idxmin()

# 20. Pivot: min and max Hue, margins
pd.pivot_table(df, index='Wine', values='Hue', aggfunc=[min, max], margins=True)
